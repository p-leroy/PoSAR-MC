```python
def readHeader(filename):
    params = {}
    with open(eFile) as f:
        # Ant32 Data File V1.0
        params['version'] = f.readline().rstrip()
        # Date=20/04/2017
        params['date'] = f.readline().split("=")[1].rstrip()
        # Operator=FB
        params['operator'] = f.readline().split("=")[1].rstrip()
        # Title=SAPHIR
        params['title'] = f.readline().split("=")[1].rstrip()
        # Remarks=Freq. = 4.1 GHz
        params['Freq'] = float(f.readline().split("=")[2].split(" GHz")[0])
        # ScanAxis=azimuth
        params['ScanAxis'] = f.readline().split("=")[1].rstrip()
        # StepAxis=polar : 0 deg
        params['StepAxis'] = float(f.readline().split(" ")[2])
        # StepPosition=0
        params['StepPosition'] = float(f.readline().split("=")[1])
        # StepUnit=deg
        params['StepUnit'] = f.readline().split("=")[1].rstrip()
        # FreqValue=4.100000
        params['FreqValue'] = float(f.readline().split("=")[1])
        # RecordIncrement=1
        params['RecordIncrement'] = float(f.readline().split("=")[1])
        # XyzImgAxis=0
        params['XyzImgAxis'] = f.readline().split("=")[1].rstrip()
        # XyzImgLabel=0
        params['XyzImgLabel'] = f.readline().split("=")[1].rstrip()
        # Abscissa=deg
        params['Abscissa'] = f.readline().split("=")[1].rstrip()
        # Ordinate=dB
        params['Ordinate'] = f.readline().split("=")[1].rstrip()
        # Data#1
        params['data'] = f.readline().split("#")[1].rstrip()
    return params
```
